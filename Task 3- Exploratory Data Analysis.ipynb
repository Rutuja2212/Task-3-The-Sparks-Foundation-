```python
#Data Visualization
```

```python
import seaborn as sns
sns.pairplot(data)
```

```python
#finding the pairwise correlation between the columns and visualizing the heatmaps
data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True,cmap="Blues")
plt.show()
```

```python
#**visualizing the ship mode**
```

```python
colors=['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.figure(figsize=(6,6))
plt.title('Ship Mode')
plt.pie(data['Ship Mode'].value_counts(),labels=data['Ship Mode'].value_counts().index,autopct='%1.1f%%',colors=colors)
centre_circle=plt.Circle((0,0),0.70,fc='White')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
#Visualizing the categories
plt.figure(figsize=(6,6))
textprops={"fontsize":15}
plt.title('Category')
plt.pie(data['Category'].value_counts(),labels=data['Category'].value_counts().index,autopct='%1.1f%%',textprops=textprops,
       colors=colors)
plt.show()
```

```python
#computing top categories in terms of sales from first 100 obs.
top_category_s=data.groupby('Category').Sales.sum().nlargest(n=100)
#computing top categories in terms of profit from first 100 obs.
top_category_p=data.groupby('Category').Profit.sum().nlargest(n=100)
#ploting to see it visuailly
plt.style.use('seaborn')
top_category_s.plot(kind='bar',figsize=(10,5),fontsize=14)
top_category_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title("Top Category Sales vs Profit",fontsize=15)
plt.show()
    
```

```python
*The above clearly show depicts technology and Office Supplies have the most sales and profit.*
```

```python
#visualizing the sales profits in different sub-categories
```

```python
plt.figure(figsize=(8,8))
textprops={"fontsize":15}
plt.title("Sub-Category")
plt.pie(data["Sub-Category"].value_counts(),labels=data["Sub-Category"].value_counts().index,
       autopct='%1.1f%%',textprops=textprops)
plt.show()
```

```python
#computing top sub-categories in terms of sales from first 100 obs.
top_subcategory_s=data.groupby('Sub-Category').Sales.sum().nlargest(n=100)
#computing top sub-categories in terms of profit from first 100 obs.
top_subcategory_p=data.groupby('Sub-Category').Profit.sum().nlargest(n=100)
#ploting to see it visuailly
plt.style.use('seaborn')
top_subcategory_s.plot(kind='bar',figsize=(10,5),fontsize=14)
top_subcategory_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title("Top Sub-Category Sales vs Profit",fontsize=15)
plt.show()
    
```

```python
# A more datailed view
plt.figure(figsize=(14,12))
statewise=data.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()  #h for horizontal
```

```python
From above graph we can show that Copiers and Phones have the highest sales and profit.Tables have negative profit.
```

```python
#visualizing the sales vs profit in different regions
```

```python
plt.figure(figsize=(6,6))
plt.title("Region")
plt.pie(data["Region"].value_counts(),labels=data["Region"].value_counts().index,
       autopct='%1.1f%%',textprops=textprops)
plt.show()
```

```python
#computing top region in terms of sales from first 100 obs.
top_subcategory_s=data.groupby('Region').Sales.sum().nlargest(n=100)
#computing top region in terms of profit from first 100 obs.
top_subcategory_p=data.groupby('Region').Profit.sum().nlargest(n=100)
#ploting to see it visuailly
plt.style.use('seaborn')
top_subcategory_s.plot(kind='bar',figsize=(10,5),fontsize=14)
top_subcategory_p.plot(kind='bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Region',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title("Top Region Sales vs Profit",fontsize=15)
plt.show()
    
```

```python
The above graph show that West have more profit though sales in the East less as compared to West
```

```python
# A more detailed view
plt.figure(figsize=(14,12))
statewise=data.groupby(['State'])['Profit'].sum().nlargest(50)
statewise.plot.barh()  #h for horizontal
```

```python
The above graph show that California and New Youk have the highest sales and profit
```
